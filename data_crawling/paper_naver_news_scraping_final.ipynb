```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from pandas import DataFrame
import os

## Chrome driver 옵션 설정 ############################
options = webdriver.ChromeOptions()
# options.add_argument('window-size=360,800')
options.add_argument('--start-maximized') ## 윈도우 크기 최대화
#######################################################

## 2. 웹 브라우저 실행
## 지면 
# url = 'https://search.naver.com/search.naver?where=news&query=%EC%BD%94%EB%A1%9C%EB%82%9819&sm=tab_opt&sort=0&photo=3&field=0&pd=3&ds=2020.01.20&de=2022.01.19&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom20200120to20220119%2Ca%3Aall&is_sug_officeid=0'

## 2020.01.20. ~ 2020.12.31.
# search_date = [('2020.01.20', '2020.01.31'), 
#                ('2020.02.01', '2020.02.28'),
#                ('2020.03.01', '2020.03.15'),
#                ('2020.03.16', '2020.03.20'),
#                ('2020.03.21', '2020.03.31'),
#                ('2020.04.01', '2020.04.10'),
#                ('2020.04.11', '2020.04.21'),
#                ('2020.04.11', '2020.04.21'),
#                ('2020.04.22', '2020.04.25'),
#                ('2020.04.26', '2020.04.30'),
#                ('2020.05.01', '2020.05.15'),
#                ('2020.05.16', '2020.05.25'),
#                ('2020.05.26', '2020.05.31'),
#                ('2020.06.01', '2020.06.10'),
#                ('2020.06.11', '2020.06.20'),
#                ('2020.06.21', '2020.06.30'),
#                ('2020.07.01', '2020.07.10'),
#                ('2020.07.11', '2020.07.20'),
#                ('2020.07.21', '2020.07.31'),
#                ('2020.08.01', '2020.08.10'),
#                ('2020.08.11', '2020.08.20'),
#                ('2020.08.21', '2020.08.31'),
#                ('2020.09.01', '2020.09.10'),
#                ('2020.09.11', '2020.09.20'),
#                ('2020.09.21', '2020.09.31'),     
#                ('2020.10.01', '2020.10.10'),
#                ('2020.10.11', '2020.10.20'),
#                ('2020.10.21', '2020.10.31'),
#                ('2020.11.01', '2020.11.10'),
#                ('2020.11.11', '2020.11.20'),
#                ('2020.11.21', '2020.11.31'),
#                ('2020.12.01', '2020.12.10'),
#                ('2020.12.11', '2020.12.20'),
#                ('2020.12.21', '2020.12.31'),
#                ('2021.01.01', '2021.01.10'), 
#                ('2021.01.11', '2021.01.20'), 
#                ('2021.01.21', '2021.01.31'),
#                ('2021.02.01', '2021.02.10'),
#                ('2021.02.11', '2021.02.20'),
#                ('2021.02.21', '2021.02.28'),
#                ('2021.03.01', '2021.03.10'),
#                ('2021.03.11', '2021.03.15'),
#                ('2021.03.16', '2021.03.20'),
#                ('2021.03.21', '2021.03.31'),
#                ('2021.04.01', '2021.04.10'),
#                ('2021.04.11', '2021.04.21'),
#                ('2021.04.22', '2021.04.25'),
#                ('2021.04.26', '2021.04.30'),
#                ('2021.05.01', '2021.05.10'),
#                ('2021.05.10', '2021.05.15'),
#                ('2021.05.16', '2021.05.20'),
#                ('2021.05.21', '2021.05.31'),
#                ('2021.06.01', '2021.06.10'),
#                ('2021.06.11', '2021.06.20'),
#                ('2021.06.21', '2021.06.30'),
#                ('2021.07.01', '2021.07.06'),
#                ('2021.07.07', '2021.07.10'),
#                ('2021.07.11', '2021.07.20'),
#                 ('2021.07.21', '2021.07.31'),
#                ('2021.08.01', '2021.08.10'),
#                ('2021.08.11', '2021.08.20'),
#                ('2021.08.21', '2021.08.31'),
#                ('2021.09.01', '2021.09.10'),
#                ('2021.09.11', '2021.09.20'),
#                ('2021.09.21', '2021.09.31'),     
#                ('2021.10.01', '2021.10.10'),
#                ('2021.10.11', '2021.10.20'),
#                ('2021.10.21', '2021.10.31'),
#                ('2021.11.01', '2021.11.10'),
#                ('2021.11.11', '2021.11.20'),
#                ('2021.11.21', '2021.11.31'),
#                ('2021.12.01', '2021.12.10'),
#                ('2021.12.11', '2021.12.20'),
#                ('2021.12.21', '2021.12.31'),
#                ('2022.01.01', '2022.01.10'),
#               ]

# startDt, endDt = ('2020.06.21', '2020.06.30')

search_date = [

               ('2022.01.11', '2022.01.19'),
              ]

for startDt, endDt in search_date:
    ## 1. 크롬 드라이버
    driver = webdriver.Chrome(options=options)
    # driver = webdriver.Chrome()
    
    url = f'https://search.naver.com/search.naver?where=news&query=%EC%BD%94%EB%A1%9C%EB%82%9819&sm=tab_opt&sort=0&photo=1&field=0&pd=3&ds={startDt}&de={endDt}&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom20200120to20201231&is_sug_officeid=0'

    ##
    # url = 'https://search.naver.com/search.naver?where=news&query=%EC%BD%94%EB%A1%9C%EB%82%9819&sm=tab_tmr&nso=so:r,p:from20200120to20220119,a:all&sort=0'
    driver.get(url)
    print('1. 웹 브라우저 실행 완료됨 ---> 1초 쉬기')
    sleep(1)

    news_list = []

    for current_page in range(1, 10000):
        ##print(f'<<현재 {current_page}페이지 >>')

        ## [스크롤 이동]
        ## 1. 현재 스크롤 전체 높이
        web_page_height = driver.execute_script('return document.body.scrollHeight')

        ## 2. 페이지 하단으로 이동
        driver.execute_script(f'window.scrollTo(0, {web_page_height})')

        ## 웹 페이지 요청
        html = driver.page_source
        sleep(1)
        soup = BeautifulSoup(html, 'html.parser')

        ## 블로그 정보 추출
        news_elements = soup.find_all('div', class_='news_area')
        news_list.extend(news_elements)
        
        ## 다음 페이지 클릭
        if not soup.find('a', attrs={'class': 'btn_next'}).get('href'):
            break

        driver.find_element(By.CSS_SELECTOR, '#main_pack > div.api_sc_page_wrap > div > a.btn_next').click()
        sleep(1)

    print('<< 데이터 추출 시작 >>')
    corona_news_info = []    
    
    for news in news_list:
        press = news.find('a', class_='info press').text ## 언론사
        date = news.find('span', class_='info').text     ## 등록일
        title = news.find('a', class_='news_tit').text   ## 기사 제목
        news_preview = news.find('a', class_='api_txt_lines dsc_txt_wrap').text 

    #     print(press)
    #     print(title)
    #     print(news_preview)
        corona_news_info.append([date, press, title, news_preview])     
    

    print('<< 데이터 저장 시작 >>')   
    from pandas import DataFrame
    from datetime import datetime

    date = str(datetime.now().date()).replace('-', '')
    time = str(datetime.now().time()).replace(':', '.')

    # print(f'corona_news_info_{date}_{time}.csv')
    table = DataFrame(corona_news_info, columns=['등록일', '언론사', '기사제목', '기사내용'])
    table.to_csv(f'corona_news_info_{date}_{time}_{startDt}_{endDt}.csv', index=False, encoding='utf-8')
    
    ## txt 파일로 저장
    with open(f'corona_news_info_{startDt}_{endDt}.txt', 'w', encoding='utf-8') as file:
        file.write(str(news_list))

    print(f'corona_news_info_{startDt}_{endDt}.txt 저장 완료 !!')
    sleep(3)
    
    driver.quit()
```
